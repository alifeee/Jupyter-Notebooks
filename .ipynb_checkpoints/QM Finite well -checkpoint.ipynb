```python
# Finite well, bound states
class BoundFiniteWell:
    def __init__(self, V0, width, mass=510998.946, # V0 in [eV], width in [nm] mass in [eV / c^2]
                 energyResolution=50, spatialResolution=1000):
        # Initial: well depth, well width, particle mass (not used), 
        # how many energies to test, the spatial resolution
        import numpy as np
        from scipy.integrate import odeint, simps
        from scipy.optimize import brentq
        self.V0 = V0
        self.width = width
        self.energyResolution = energyResolution
        self.spatialResolution = spatialResolution
        self.xs = np.linspace(-width*1.5, width*2.3, self.spatialResolution)
#         self.mass = mass
#         global HBAR
#         HBAR = 658.21 # [eV s / nm^2]

    def V(self, x):
        # The potential function
        if abs(x) < self.width/2:
            return 0
        else:
            return self.V0
    def dU_dx(self, U, x, E):
        # Return a vector of [', '']
        return [U[1],
                2*U[0]*(self.V(x)-E)
#                 2*self.mass/HBAR*U[0]*(self.V(x)-E)
               ]
    def psi(self, energy):
        # Returns the height of psi for an arbitrary energy
        Us = odeint(lambda U, x: self.dU_dx(U, x, energy),
                    [0.00001,0], self.xs
                   )
        return Us[:, 0]
    def psiFromCentre(self, energy, initial):
        newXs = self.xs[self.xs > 0]
#         print(initial)
        Us = odeint(lambda U, x: self.dU_dx(U, x, energy),
                    initial, newXs
                   )
        return Us[:, 0]
        
    def psiNormalised(self, energy):
        # Returns the normalised wave function
        # NOTE: Only works properly for allowed energy states
        psi = self.psi(energy)
        integral = simps(np.power(psi, 2), x=self.xs)
        return psi / np.sqrt(integral)
        
    def findEnergiesNumerically(self):
        # Goes through N energies, finds ones where (b) = 0, b > width
        allowedEnergies = []
        allEnergies = np.linspace(0, self.V0, self.energyResolution)
        lastValues = [self.psi(energy)[-1] for energy in allEnergies]
        for i in range(len(lastValues)-1):
            if np.sign(lastValues[i]) != np.sign(lastValues[i+1]):
                newEnergy = brentq(lambda energy: self.psi(energy)[-1],
                                   allEnergies[i], allEnergies[i+1]
                                  )
                allowedEnergies.append(newEnergy)
        return allowedEnergies
        
    
    
well = BoundFiniteWell(20, 2)

for i, energy in enumerate(well.findEnergiesNumerically()):
    initial = [(-1)**(int(i/2))    *((i+1)%2) + 0.0001,
               (-1)**(int((i+1)/2))*(i%2)
              ] # creates [1,0], [0,-1], [-1,0], [0,1] in sequence
    print("E = ", energy)
    psis = well.psi(energy)
    psis2  = well.psiFromCentre(energy, initial)
    plt.plot(well.xs, psis/max(psis))
#     plt.show()
    trunc = 200
    plt.plot(well.xs[well.xs>0][:trunc], psis2[:trunc]/max(np.abs(psis2[:trunc])))
    plt.show()
    

```
